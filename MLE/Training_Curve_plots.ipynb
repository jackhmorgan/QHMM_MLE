```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
with open('ClassicalConvergence/qhmm_test (3).json', 'r') as file:
    data = json.load(file)
```

```python
num_time_steps = 100
trial_data = data[str(num_time_steps)]
gen_likelihood = trial_data['gen_model_likelihood']
training_curve = [-likelihood for likelihood in trial_data['likelihood_curve'] if not likelihood==0]
```

```python
iteration = np.linspace(0, len(training_curve), len(training_curve))
plt.plot(iteration,training_curve, label='Trained Model Likelihood')
plt.hlines(-gen_likelihood, min(iteration), max(iteration), colors='r', label='Generating Model Likelihood')
plt.legend()
plt.xlabel('Training Iteration')
plt.ylabel('Negative Log Likelihood')
title = 'Training Curve for QHMM'
plt.title(title)
plt.savefig(title+'.png')
```
