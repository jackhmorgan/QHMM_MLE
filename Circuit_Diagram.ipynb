```python
from HMM import QHMM
from HMM.utils.qhmm_utils import result_getter, qiskit_simulator_result_getter
from qiskit.circuit.library import RealAmplitudes
from qiskit.circuit import ParameterVector
from qiskit import QuantumCircuit
```

```python
initial_state = QuantumCircuit(1)
initial_state.h(0)

ansatz = RealAmplitudes(num_qubits=3, reps=1)
```

```python
class circuit_diagram_result_getter(qiskit_simulator_result_getter):
    def generate_sequence(self, circuit):
        return circuit.decompose(reps=2).draw('mpl')     
```

```python
theta = ParameterVector(name='Theta', length=ansatz.num_parameters)
```

```python
result_getter = circuit_diagram_result_getter()
example = QHMM(result_getter=result_getter,
               initial_state=initial_state,
               ansatz=ansatz,
               theta=theta)
```

```python
example.generate_sequence(3)
```
